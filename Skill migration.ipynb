#### Findings:

skill_raw_data dataframe contains 20647 rows and 29 columns which is incorrect as per the skill_raw_data.info() method it should contain 17617 rows and 12 columns. Hence, we should start cleaning the data. 

```python
# Remove missing values. Drop the rows where all elements are missing.
skill_data_clean = skill_raw_data.dropna(how="all") 
skill_data_clean
```

```python
 # Remove missing values. Drop the columns where at least one element is missing.
skill_dataframe = skill_data_clean.dropna(axis='columns') 
skill_dataframe
```

```python
# FinTech, Mobile Application Development, Development Tools are the gaining skills of 2015 and 2019. 
# Let's find out the trend FinTech, Mobile Application Development, Development Tools and Artificial Intelligence (AI) from 2015 to 2019. 
# Hence, we need a dataframe which contain skill_group_name of UK and net_per_10K from 2015 to 2019.
UK_skill_group_name = pd.DataFrame({'skill_group_name': UK.skill_group_name ,'net_per_10K_2015': UK.net_per_10K_2015,'net_per_10K_2016': UK.net_per_10K_2016,'net_per_10K_2017': UK.net_per_10K_2017,'net_per_10K_2018': UK.net_per_10K_2018,'net_per_10K_2019': UK.net_per_10K_2019})
UK_skill_group_name
```

```python
# We need to set the index to skill_group_name.  
UK_skill_group_name.set_index('skill_group_name')
```

```python
# Transpose the UK_skill_group_name in order to create the graph for FinTech, Mobile Application Development, Development Tools.
UK_Tranpose = UK_skill_group_name.set_index('skill_group_name')
UK_skill_trend = UK_Tranpose.T
UK_skill_trend
```

```python
# Let's visualize the trend FinTech, Mobile Application Development, Development Tools and Artificial Intelligence (AI) from 2015 to 2019. 
plt.figure(figsize=(12,8))
FinTech = UK_skill_trend['FinTech'].plot.line(color='r')
Mobile_Application_Development = UK_skill_trend['Mobile Application Development'].plot.line(color='g')
Development_Tools = UK_skill_trend['Development Tools'].plot.line(color='b')
Artificial_Intelligence = UK_skill_trend['Artificial Intelligence (AI)'].plot.line(color='y')
plt.legend(loc='upper center')
plt.xlabel('Year')
plt.ylabel('Frequency')
```

#### Findings:

As per the graph, we can see that Artificial Intelligence (AI) is the most gaining skills from 2015 to 2019 among FinTech, Mobile Application Development and Development Tools. However, in the year 2017 all these skills have fallen. 

In the year 2019, we can see that gaining skills, Artificial Intelligence (AI)  and Development Tools was almost same as it was in the year 2015 whereas FinTech was slightly more. However, Mobile Application Development was not as it was in the year 2015.
